##1. and 2.

```python
# copy in provided matrices as numpy arrays
matrix_A = np.array([[1, -2, 3, 7],[2, 1, 1, 4],[-3, 2, -2, 10]])
matrix_B = np.array([[0, 1, -3, -2],[10, -1, 2, -3],[5, 1, -1, 4]])
matrix_C = np.array([[4, 0, -2, 3],[3, 6, 9, 7],[2, 2, 5, 1],[9, 4, 6, -2]])

# Print the dimensions of the given matrices using .shape
print ('A dimensions',matrix_A.shape)
print ('B dimensions',matrix_B.shape)
print ('C dimensions',matrix_C.shape)
```

```python
# Determine if the following matrix operations are defined or undefined.
# A + B is defined
print("matrix_A + matrix_B =")
print(matrix_A + matrix_B)
```

##c)

```python
# A * C is defined
print("matrix_A * matrix_C =")
print(matrix_A @ matrix_C)
```

##d)

```python
# C^T is defined
print("matrix_C transpose =")
print(matrix_C.T)
```

##4.

```python
# Illustrate the following property of matrix transpositions
# Compute the left-hand side of the property (A + B)^T
# First add matrices A and B, then take the transpose of the result
left = (matrix_A + matrix_B).T
#Compute the right-hand side of the property A^T + B^T
# Take the transpose of each matrix first, then add them together
right = (matrix_A.T + matrix_B.T)

# Print the results to compare the two
print(left)
print(right)
```

##5.


```python
#Illustrate the following property of matrix transposition
# Compute the left-hand side of the equation (AC)^T
# First multiply A and C, then take the transpose of the result
left = (matrix_A @ matrix_C).T

#Compute the right-hand side of the equation C^T A^T
# First take the transpose of C and A individually
# Then multiply them in reverse order (C^T followed by A^T)
right = (matrix_C.T @ matrix_A.T)

# Print both matrices and see that they are equal to eachother
print(left)
print(right)
```

##6.

```python
#Inverse of matrix_C
# np.linalg.inv() is a NumPy function that returns the inverse of a square matrix
matrix_C_inverse = np.linalg.inv(matrix_C)

print("matrix_C_inverse =", matrix_C_inverse)
```
