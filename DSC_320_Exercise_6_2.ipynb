##Problem 1.

```python
#Read csv and assign as data frame
df = pd.read_csv("/content/drive/MyDrive/calif_housing_data.csv")

# clean data to remove the rows that have a na value
indices_list = df["total_bedrooms"].index[df["total_bedrooms"].isna()].tolist()
df = df.drop(indices_list)

# Define the normalize function that puts all of the features values on a scale between 0 to 1.
# convert input to NumPy array
# calculate the sample mean
# calculate sample standard deviation (n-1)
def normalize(x):
  x = np.array(x)
  min_val = np.min(x)
  max_val = np.max(x)

  return (x-min_val)/ (max_val - min_val)

# Print normalized data using a column from the df
normalized = normalize(df["housing_median_age"])
print ("normalized:", normalized)
```

## Problem 2.

```python
# Create the standardize function
# Commented out my debugging when I was getting nan in the last problem.
def standardize(x):
  # print(x.head(10))
  x_array = np.array(x)
  # has_nan = np.isnan(x_array).any()
  # print(f"{x_array=}, {len(x_array)=}, {has_nan=}")
  samp_mean = np.mean(x_array)
  stan_dev = np.std(x_array, ddof = 1)  # Find standard deviation of a sample by setting the delta degrees of freedom to 1
  # print(f"{samp_mean=} {stan_dev=}")
  return (x-samp_mean)/stan_dev
```
