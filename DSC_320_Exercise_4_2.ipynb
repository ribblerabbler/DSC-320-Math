## Calculating Limits Numerically

```python
# Import necessary libraries
import numpy as np
```

```python
# Evaluate the function f at appropriate values to calculate lim f(x) as x approaches 3
#vals = [2.5,2.7,2.9, 2.99, 2.999]
vals = np.geomspace(2.5, 3, num=25)

# define the provided function f(x)
def f(x):
  return 2*x**3 - 4*x +1

# Call the function to find values as x approaches 3
for x in vals:
  new_val = f(x)
  print(f"input val = {x}, new val = {new_val}")
```

```python
# Evaluate values of x approaching 0 to approximate the limit of g(x)
# np.logspace generates values between 10^-1 and 10^0 (0.1 to 1)
vals = np.logspace(-1,0,10)
#Define the provided function g(x)
def g(x):
  return (np.e**x-1)/x

# Call the fuction to loop through each value of x
# Calculate then print the input and new values.
for x in vals:
  new_val = g(x)
  print(f"input val = {x}, new val = {new_val}")
```

For part (a), evaluating values close to 3 shows that
f(x) approaches 43. Sincef(3)=43, the limit can be found by direct substitution, indicating the function is continuous at that point.

For part (b), evaluating values close to 0 shows that g(x) approaches 1. However, plugging in x=0 results in an undefined expression (0/0), so the limit cannot be found by direct substitution. Although the limit exists and equals 1, the function is not defined at x=0.

Therefore, f(x) is continuous at x=3, while
g(x) would only be continuous at x=0 if it were redefined so that g(0)=1.Continuity requires both the existence of the limit and that it equals the function's value at that point.

## Average Rate of Change

```python
# Define the function f(x) takes in a mathematical function, e.g., f(x) = 3x^2 and two numbers a and b
# and returns the average rate of change of the function between a and b.
def f(x):
  return 3 * (x ** 2)

# Define the function that takes two values and a function f(x)= 3x^2 to calculate the Average Rate of Change
def aroc(a, b, func):
  numerator = func(b) - func(a)
  denominator = b - a
  return numerator / denominator
```

```python
# Random a and b values to calculate the rate of change
a_vals = [ 0,   1,  2,  3,   4,  8]
b_vals = [122, 11,  5,  4, 999, 50]

# This for loop calculates the rate of change between each a and b value.
for i in range(len(a_vals)):
  a = a_vals[i]
  b = b_vals[i]
  roc = aroc(a, b, f)
  print(f"a = {a}, b = {b}, average rate of change = {roc}")
```
