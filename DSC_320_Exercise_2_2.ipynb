## Problem 5


```python
import matplotlib.pyplot as plt

# Code to add line to graph from problem 1
# I adjusted the slpe and y-intercept starting with -0.05 slope and 100 intercept until the line ran through the majority of the data points.
slope = -0.01
intercept = 55

# Calculation for finding slope y = mx+b
x = df['weight']
y = slope * x + intercept

# Create the scatterplot with a line of best fit.
plt.scatter(df['weight'], df['hwy_mpg'])
plt.plot(x, y, color='blue')
plt.show()
```

## Problem 6

```python
# Use Python to find the best-fit (least-squares) line. The Scikit-learn package is a good choice to use for this.
# Import necessary tools
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Assign the variables from the data frame
x = df['weight']
y = df['hwy_mpg']

# Use polyfit to find the line of best fit with least-squares
slope, intercept = np.polyfit(x, y, 1)

# Read out the calculated values above the graph
print('slope', slope)
print('intercept', intercept)

# Create the scatterplot
plt.scatter(df['weight'], df['hwy_mpg'])
plt.plot(x, slope * x + intercept, color='red')

# Lable the graph
plt.xlabel("Weight (pounds)")
plt.ylabel("Highway MPG")
plt.title("Best-Fit Line")

# Show the graph
plt.show()
```

## Problem 7

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

# My manual adjustments to the scatter plot and best fit line
manual_slope = -0.01
manual_intercept = 60

# Use y = mx+b to find the manual prediction of y
y_pred_manual = manual_slope * x + manual_intercept

# Calculate RMSE
rmse_manual = np.sqrt(np.mean((y - y_pred_manual) ** 2))

print("Manual RMSE:", rmse_manual)

# Calculate the slope
y_predicted = slope * x + intercept
y_ = df['hwy_mpg']

# Manual Interpretation for Calculating the Root Mean Squared Error us NumPy
calculated_RMSE = np.sqrt(np.mean((y_ - y_predicted)**2))
print("Calculated RMSE:", calculated_RMSE)
```
