```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rt_riemann_sum(f,a,b,n):  # define function to find the right riemann sum
  dx = (b-a)/n                # width of the rectangles
  x = np.arange(a+dx,b+dx,dx) # right endpoints so must start at a+dx and go to b
  return np.sum(f(x) * dx)

print(f"right riemann sum = {rt_riemann_sum(f,a,b,n)}")

# Math Check
#f = lambda x: x**2
#print(rt_riemann_sum(f, 0, 1, 1000))  # should be close to 1/3
```

## a)

```python
# Interval and number of rectangles
a = 0
b = 2
n = 4
dx = (b - a) / n

# Smooth curve for the function
x_curve = np.linspace(a, b, 400)
y_curve = f(x_curve)

# Right endpoints for rectangles
x_right = np.arange(a + dx, b + dx, dx)

# Plot the curve
plt.plot(x_curve, y_curve, label='f(x) = sqrt(4 - x^2)')

# Plot rectangles
for x in x_right:
    left = x - dx
    height = f(x)

 # Draw rectangles
    plt.bar(left, height, width=dx, align='edge', alpha=0.3, edgecolor='black')

# Labels and title
plt.title('Right Riemann Sum (n=4) for f(x) = sqrt(4 - x^2)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.show()
```

##b)

```python
# Define the given function
def f(x):
   return np.sqrt(4-(x**2))

# Call the function to find the right riemann sum of f(x) = sqrt(4-x**2)
R4 = rt_riemann_sum(f,0,2,4)
print("R4 =", R4)
```

##c)

```python
# Calculate right Riemann sums to find the area under the curve.
# Use increasing values for n to approximate the definite integral.
for n in [4, 10, 50, 100, 500, 1000]:
  approximate = rt_riemann_sum(f,0,2,n)
  print (f"R_{n} = {approximate}") # values approach pi

# Print the exact value for comparison
# Math check
print("Exact area =", np.pi)
```

The value of this area is pi or the area of a quarter circle of radius 2.
