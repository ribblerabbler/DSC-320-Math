```python
# Read in the mushroom data frame.
df = pd.read_csv("/content/drive/MyDrive/mushroom_data.csv")

# Assign the variables from the df using the column names.
actual = df["actual"]
predicted = df["predicted"]

# Define the function that calculates the accuracy of the mushroom data frame.
# The values are p for poisonous and e for edible.
# This functions counts the correct predictions by comparing the actual and predicted value.
# If the predicted values matches the actual it adds 1 to the correct_predictions variable.
# The loop continues until it has read through all of the data entries.
# Then it returns the correct__predictions total.
def calc_accuracy(actual,predicted):
  correct_predictions = 0
  for i in range(len(actual)):
    if actual[i] == predicted[i]:
      correct_predictions += 1
  return correct_predictions

#Call the function to carry out the function.
calc_accuracy(actual,predicted)
```
